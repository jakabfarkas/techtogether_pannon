```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
```

```python
data = pd.read_excel('2_melleklet_EventLog_Pannon Egyetem.xlsx', index_col=None)
```

```python
data = np.array(data)
            
for i in range(len(data)):
    for j in range(i+1, len(data)):
        if (i < len(data) and j < len(data) and
            data[i][0] == data[j][0] and
            data[i][4] == data[j][4] and
            data[i][5] == data[j][5] and
            data[i][6] == data[j][6]):
            data[i][1] += data[j][1]
            data[i][2] += data[j][2]
            data = np.delete(data, j, 0)
```

```python
for i in range(len(data)):
    nok = nok + data[i][2]
```

```python
workorders = []
op_cnt = []
time = []
n_ok = []
tech = []
work = []
us = []
for i in range(len(data)):
    if data[i][0] not in workorders:
        time_min = data[i][3]
        time_max = data[i][3]
        nok = 0
        cnt = 0
        technologies = []
        workplaces = []
        users = []
        for j in range(i, len(data)):
            if (data[j][0] == data[i][0]):
                if (data[j][3] < time_min):
                    time_min = data[j][3]
                if (data[j][3] > time_max):
                    time_max = data[j][3]
                nok = nok+data[j][2]
                cnt = cnt+1
                if (data[j][4] not in technologies):
                    technologies.append(data[j][4])
                if (data[j][5] not in workplaces):
                    workplaces.append(data[j][5])
                if (data[j][6] not in users):
                    users.append(data[j][6])
        print(data[i][0])
        workorders.append(data[i][0])
        print('   Operations: ', cnt)
        op_cnt.append(cnt)
        print('   Time from first to last: ', time_max-time_min)
        time.append((time_max-time_min).total_seconds())
        print('   NOK in all operations: ', nok)
        n_ok.append(nok)
        print('   Technologies: ', len(technologies))
        tech.append(len(technologies))
        print('   Workplaces: ', len(workplaces))
        work.append(len(workplaces))
        print('   Users: ', len(users))
        us.append(len(users))
        print('********************')
        
plt.bar(workorders, op_cnt, 1)
plt.xlabel('Workorders')
plt.ylabel('Operations')
plt.show()
plt.bar(workorders, time, 1)
plt.xlabel('Workorders')
plt.ylabel('Time from first to last')
plt.show()
plt.bar(workorders, n_ok, 1)
plt.xlabel('Workorders')
plt.ylabel('NOK')
plt.show()
plt.bar(workorders, tech, 1)
plt.xlabel('Workorders')
plt.ylabel('Technologies')
plt.show()
plt.bar(workorders, work, 1)
plt.xlabel('Workorders')
plt.ylabel('Workplaces')
plt.show()
plt.bar(workorders, us, 1)
plt.xlabel('Workorders')
plt.ylabel('Users')
plt.show()
```
